```python
# 메뉴 정보
item = {'자장면': 5000, '짬뽕' : 6000, '볶음밥' : 6000, '탕수육' : 10000}
# 테이블 별 주문 정보
table_list = {}
# 현재 주문 정보
order_list = {}
# 관리자 비밀번호
password = '0000'
```

```python
# 사용자 모드 선택
def mode():
    login = int(input('모드 선택 (0:관리자, 1:주문, 2:결제, 3:프로그램 종료) : '))

    # [0 = 관리자, 1 = 주문 화면, 2 = 결제 화면]
    if login == 0:
        dev_op_check()
    elif login == 1:
        order_table()
    elif login == 2:
        payment()
    elif login == 3:
        print('---- 종료되었습니다 ---')
    else:
        print('선택이 잘못되었습니다. 다시 선택하세요.')
        mode();
```

```python
# 관리자 비밀번호 확인
def dev_op_check():
    check = input('(관리자 페이지) 비밀번호를 입력하세요 : ')
    if password == check:
        dev_op()
    else:
        print('비밀번호가 틀립니다.')
        mode()
        
# 메뉴 관리 기능 [0: 메뉴 추가, 1: 메뉴 삭제, 2: 메뉴 수정, 3: '사용자 모드 설정'으로 이동]
def dev_op():
    setting = int(input('메뉴 관리 (0: 추가, 1: 삭제, 2: 수정, 3: 나가기) : '))
    
    if setting == 0:
        menu_add()
    elif setting == 1:
        menu_del()
    elif setting == 2:
        menu_fix()
    elif setting == 3:
        mode()
    else:
        print('입력값 오류입니다. 다시 선택하세요.')
        dev_op()
        
# 메뉴 추가
def menu_add():
    global item
    add_menu_name = input('추가할 메뉴를 입력하세요 (0: 뒤로 가기) : ')
    if add_menu_name == '0':
        pass
    elif add_menu_name in item:
        print('해당 메뉴는 이미 있습니다!')
    else:
        price = int(input(f'{add_menu_name} 가격 설정 : '))
        item[add_menu_name] = price
        print(f'{add_menu_name}의 메뉴가 {price}원에 추가되었습니다!\n')
    dev_op()
        
        
# 메뉴 수정
def menu_fix():
    global item
    print('\n--------메뉴 정보--------')
    for key, value in item.items():
        print(f'{key} : {value}원')
        
    fix_item = input('\n수정할 메뉴를 선택하세요(0: 뒤로 가기) : ')
    if fix_item == '0':
        dev_op()
    elif fix_item in item:
        fix_type = int(input('1: 이름 수정, 2: 가격 수정, 3: 취소 : '))
        if fix_type == 1:
            fix_name = input('수정할 메뉴 이름을 작성하세요 : ')
            item[fix_name] = item[fix_item]
            del item[fix_item]
            print(f'수정이 완료되었습니다. {fix_name} = {item[fix_name]}')
        elif fix_type == 2:
            fix_value = int(input(f'{fix_item}의 가격을 설정하세요 : '))
            item[fix_item] = fix_value
            print(f'수정이 완료되었습니다. {fix_item} = {fix_value}')
        elif fix_type == 3:
            pass
        else:
            print('잘못된 선택입니다. 메인으로 돌아갑니다.')
        dev_op()
    else:
        print('없는 메뉴입니다.')
        dev_op()

        
# 메뉴 삭제
def menu_del():
    global item
    print('\n--------메뉴 정보--------')
    for key, value in item.items():
        print(key, ':', value, '원')
    del_item = input('\n삭제할 메뉴를 선택하세요(0: 뒤로 가기) : ')
    if del_item == '0':
        pass
    elif del_item in item:
        del item[del_item]
        print(f'{del_item}의 메뉴가 삭제되었습니다.')
    else:
        print('없는 메뉴입니다.')
    dev_op()
```

```python
# 주문 할 테이블 선택
def order_table():
    table_num = int(input('테이블 번호를 입력하세요 : '))
    
    # 테이블 중복 주문 방지
    if table_num in table_list:
        print('자리가 있는 테이블 번호입니다.')
        order_table()
    else:
        print('\n--------메뉴 정보--------')
        for key, value in item.items():
            print(f'{key} : {value}원')
        order_menu(table_num)
        
# 메뉴 선택
def order_menu(table_num):
    global order_list
    
    menu_add = input('\n메뉴를 입력하세요(0 입력 시 주문 종료) : ')
    if menu_add in item:
        order_list[menu_add] = int(input('주문 개수를 입력하세요 : '))
        print('메뉴가 추가되었습니다.')
        order_menu(table_num)
    elif menu_add == '0':
        # 주문 내용이 없는 경우 테이블 생성하지 않음
        if not order_list:
            print('주문 목록이 없으므로 결제를 취소합니다.\n')
            mode()
        else:
            # 주문 정보를 table_list에 저장 Ex >> [1:{자장면 : 3}]
            table_list[table_num] = order_list
            print(f'{table_num}번 테이블 주문이 완료되었습니다!\n')
            order_list = {}
            mode()
    else:
        print('없는 메뉴입니다. 다시 입력하세요.\n')
        order_menu(table_num)
```

```python
# 결제 테이블 선택
def payment():
    num = int(input('결제할 테이블 번호를 입력하세요 : '))
    
    if num in table_list:
        print(f'\n{num}번 테이블 주문 정보\n------------------')
        total_payment = 0
        for key, value in table_list[num].items():
            print(f"{key} ({item[key]}원) : {value} = {item[key] * value}원")
            total_payment = total_payment + item[key] * value
        print('----------------------')
        print(f'총 결제금액 : {total_payment}원\n\n')
        
        paytype(num, total_payment)
    else:
        print('없는 테이블 정보입니다.')
        mode()

        
# 결제 방식 선택
def paytype(num, payment):
    global order_list
    selt = int(input('결제 방식을 선택하세요(1: 카드, 2: 현금, 3: 메인으로) : '))
    if selt == 1:
        print('결제가 완료되었습니다. 이용해 주셔서 감사합니다.')
        
        paylog('카드', payment) # 결제 로그 남기기
        
        del table_list[num]
        mode()
    elif selt == 2:
        paymoney = int(input('현금을 지불하세요 : '))
        paymoney = paymoney - payment
        
        # 지불한 금액 확인 >> 부족할 경우, 결제창으로 복귀
        if paymoney >= 0:
            print(f'현금 잔액 반환 : {paymoney}')
            print('결제가 완료되었습니다. 이용해 주셔서 감사합니다.')
            
            paylog('현금', payment)# 결제 로그 남기기
            
            
            del table_list[num]
            mode()
        else:
            print('결제 금액이 부족합니다. 결제 화면으로 돌아갑니다.')
            paytype(num, payment)
    elif selt == 3:
        mode()
    else:
        print('결제할 수 없는 수단입니다. 다시 선택하세요')
        paytype()
        
# 결제 로그 저장
def paylog(paytype, payment):
    with open("list_pay.txt", 'a') as f:
        f.write(f"결제 방식 : {paytype}, 결제 금액 : {payment}\n")
        
```
